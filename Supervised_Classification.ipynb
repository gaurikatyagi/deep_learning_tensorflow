## Logistic Regression

In a logistic classification, the output y, is the probability of the input falling under one of the classes. One probability needs to be the highest and all others need to be as close to 0 as possible.

## Softmax Function

We find these probabilities using the Softmax function on the outputs of the logistic regression:

Y = Wx + b

Where: Y: ouyput, W:weights, b:biases

Softmax Formula is:

$$S(yi)= e^{yi}/\sum_{j}^\ e^{yi}$$

here, yi are the scores of logistic regression, i.e. these are the logits for each respective xi.

The softmax function gives out proper probailities, i.e. they sum up to 1.


```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)
```

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x)/np.sum(np.exp(x), axis = 0)
```

```python
# Plot softmax curves
x = np.arange(-3.0, 8.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.7 * np.ones_like(x)])
smax = softmax(scores)
plot_object = plt.plot(x, smax.T, linewidth=2)
plt.xlabel("x: input value")
plt.ylabel("y: softmax value")
plt.legend(iter(plot_object), ("x", "feed_1", "feed_2"), 
           bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.show()
```

Let us take a sample scores input for the softmax function and see what the inputs and outputs are:

```python
scores = [[8.9, 4, 7.3], [2.9, 4.8, 0.3]]
sfmax = softmax(scores)
print sfmax
```

You can see that the softmax was calculated based on the imputs, the highest number got the highest probability. When we are doing this for logistic regression, we not only want the probabilities, we can to categorize the input to one of the three output classes (here our scores).
For this, we would require the higherst probability number to get the label 1 and all other classes to get a label 0. Also known as One-hot-encoding.

For the above example: [1 0 0]

```python
one_hots = np.zeros_like(sfmax)
one_hots = (sfmax == sfmax.max(axis=1)[:,None]).astype(int)
print one_hots
```
